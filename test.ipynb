```python
import random

def personality_test():
    """Genera un test de personalidad básico para derivación."""
    print("Test de Personalidad para Derivación al Profesional de Psicología\n")
    questions = [
        "¿Te sientes a menudo preocupado/a o ansioso/a?",
        "¿Tienes dificultades para controlar tus emociones?",
        "¿Te resulta difícil relacionarte con los demás?",
        "¿Experimentas cambios de humor repentinos?",
        "¿Te sientes abrumado/a por el estrés?",
        "¿Has perdido interés en actividades que antes disfrutabas?",
        "¿Tienes problemas para dormir o comes en exceso/demasiado poco?",
        "¿Te sientes desesperanzado/a o sin valor?",
        "¿Has tenido pensamientos de hacerte daño o dañar a otros?",
        "¿Evitas situaciones sociales o te aíslas?"
    ]
    scores = []
    print("Responde con 'Sí' o 'No'.")
    for i, q in enumerate(questions):
        while True:
            answer = input(f"{i+1}. {q} ").strip().lower()
            if answer in ['si', 'no']:
                scores.append(1 if answer == 'si' else 0)
                break
            else:
                print("Respuesta inválida. Por favor, responde con 'Sí' o 'No'.")

    total_score = sum(scores)
    print("\nResultados del Test de Personalidad:")
    print(f"Puntuación total: {total_score}")

    if total_score >= 5:
        print("Basado en tus respuestas, sería beneficioso que consultaras con un profesional de la psicología para una evaluación más detallada.")
    else:
        print("Tus respuestas sugieren que no hay una necesidad inmediata de derivación, pero si sientes que necesitas apoyo, no dudes en buscar ayuda profesional.")

def motivational_test():
    """Genera un test motivacional básico."""
    print("\nTest Motivacional\n")
    questions = [
        "¿Qué te impulsa a levantarte cada mañana?",
        "¿Qué metas te has propuesto alcanzar?",
        "¿Qué te apasiona y te da energía?",
        "¿Qué obstáculos has superado para lograr tus objetivos?",
        "¿Qué te motiva a seguir adelante cuando las cosas se ponen difíciles?",
        "¿Qué tipo de reconocimiento o recompensa te resulta más satisfactorio?",
        "¿Cómo te mantienes enfocado/a en tus objetivos a largo plazo?",
        "¿Qué te inspira a dar lo mejor de ti?",
        "¿Qué papel juegan tus valores personales en tu motivación?",
        "¿Cómo manejas el fracaso o la decepción en relación con tus metas?"
    ]
    answers = {}
    print("Por favor, responde honestamente a las siguientes preguntas:")
    for i, q in enumerate(questions):
        answer = input(f"{i+1}. {q}\nRespuesta: ").strip()
        answers[q] = answer

    print("\nRespuestas del Test Motivacional:")
    for q, a in answers.items():
        print(f"- {q}\n  Respuesta: {a}")

    print("\nReflexiona sobre tus respuestas para comprender mejor qué te motiva.")

def risk_tolerance_test():
    """Genera un test básico de tolerancia al riesgo."""
    print("\nTest de Tolerancia al Riesgo\n")
    questions = [
        "Prefiero una inversión con un rendimiento seguro y bajo que una con alto rendimiento y alto riesgo.",
        "Me siento cómodo/a tomando decisiones importantes sin tener toda la información.",
        "Disfruto probando cosas nuevas, incluso si hay una posibilidad de fracaso.",
        "Me preocupa perder dinero en inversiones.",
        "Soy propenso/a a asumir riesgos calculados para obtener una recompensa mayor.",
        "Evito situaciones que considero peligrosas o inciertas.",
        "Me siento bien al salir de mi zona de confort.",
        "Prefiero la estabilidad y la previsibilidad a la emoción y la aventura.",
        "Estoy dispuesto/a a arriesgarme si creo que la recompensa potencial es significativa.",
        "Me considero una persona cautelosa."
    ]
    scores = []
    print("Evalúa cada afirmación en una escala del 1 al 5, donde 1 es 'Totalmente en desacuerdo' y 5 es 'Totalmente de acuerdo'.")
    for i, q in enumerate(questions):
        while True:
            try:
                score = int(input(f"{i+1}. {q} (1-5): ").strip())
                if 1 <= score <= 5:
                    scores.append(score)
                    break
                else:
                    print("Entrada inválida. Por favor, introduce un número del 1 al 5.")
            except ValueError:
                print("Entrada inválida. Por favor, introduce un número.")

    total_score = sum(scores)
    print("\nResultados del Test de Tolerancia al Riesgo:")
    print(f"Puntuación total: {total_score}")

    if total_score >= 40:
        print("Tu puntuación sugiere una alta tolerancia al riesgo.")
    elif total_score >= 25:
        print("Tu puntuación sugiere una tolerancia moderada al riesgo.")
    else:
        print("Tu puntuación sugiere una baja tolerancia al riesgo.")

# Ejecutar los tests
personality_test()
motivational_test()
risk_tolerance_test()
```
